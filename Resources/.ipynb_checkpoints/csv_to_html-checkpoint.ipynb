```python
import os
import pandas as pd
```

```python
file_to_load=("cities.csv")
csv = pd.read_csv(file_to_load)
cities = pd.DataFrame(csv)
cities.index.name="City ID"
cities
```

```python
cities_html=cities.to_html()
print(cities_html)
```

```python
text_file = open("cities.html", "w")
text_file.write(cities_html)
text_file.close()
```
